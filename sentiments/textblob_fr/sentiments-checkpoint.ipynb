# Sentiment Analysis


```python
from textblob import TextBlob
import pandas as pd
import glob
from textblob_fr import PatternTagger, PatternAnalyzer
from nltk import sent_tokenize
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (34,16) 
```

```python
files = glob.glob("plain/*.txt") 

# create document list:
documents = []
for filename in files:
    with open(filename, 'r', encoding="utf-8", errors='ignore') as f:
        documents.append(f.read())
```

Writing a function to get the Polarity of a text:

```python
 def getPolarity(text):
   return TextBlob(text).sentiment.polarity

```

```python
Blobs = [TextBlob(document).sentiment
            for document in documents]  # turning them into Textblob-Objects
```

```python
for document in documents:
    getPolarity(document)  
```

What are the polarities of the texts in the current corpus? 

```python
Blobs
```

```python
pd.Series(Blobs)
```

```python
Sentiments = [document.polarity
                 for document in Blobs]
```

```python
pd.Series(Sentiments, index=files).plot(kind='bar')  # plotting the sentiments of the texts in the current corpus
```

```python
pd.Series(Sentiments).rolling(window=20).mean().plot()
```
